```python
x = data.drop(columns=['target'], axis=1)
```

```python
x.head()
```

```python
y = data['target']
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=2, stratify=y
)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(x_train_scaled, y_train)
```

```python
train_pred = model.predict(x_train_scaled)
test_pred = model.predict(x_test_scaled)
```

```python
print("Training accuracy:", accuracy_score(y_train, train_pred))
print("Test accuracy:    ", accuracy_score(y_test, test_pred))
print("\nConfusion matrix (test):\n", confusion_matrix(y_test, test_pred))
print("\nClassification report (test):\n", classification_report(y_test, test_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model = DecisionTreeClassifier(random_state=2)
dt_model.fit(x_train_scaled, y_train)
```

```python
dt_train_pred = dt_model.predict(x_train_scaled)
dt_test_pred = dt_model.predict(x_test_scaled)
```

```python
print("Training data shape:", x_train.shape, y_train.shape)
print("Testing data shape:", x_test.shape, y_test.shape)
```

```python
print("Decision Tree Training Accuracy:", accuracy_score(y_train, dt_train_pred))
print("Decision Tree Test Accuracy:    ", accuracy_score(y_test, dt_test_pred))
print("\nConfusion Matrix (test):\n", confusion_matrix(y_test, dt_test_pred))
print("\nClassification Report (test):\n", classification_report(y_test, dt_test_pred))
```
